# Batch Process Images

This notebook runs the complete MagicFace pipeline (preprocess, retrieve_bg, and inference) on 50 images from `processed_identities`.

```python
import os
import sys
import subprocess

# Add project root to path if needed
PROJECT_ROOT = os.getcwd()
if PROJECT_ROOT not in sys.path:
    sys.path.append(PROJECT_ROOT)

PYTHON_PATH = 'C:/Users/faris3/AppData/Local/miniconda3/envs/magicface/python.exe'
INPUT_DIR = os.path.join('test_images', 'processed_identities')
OUTPUT_DIR = 'edited_images'
os.makedirs(OUTPUT_DIR, exist_ok=True)

image_files = [f for f in os.listdir(INPUT_DIR) if f.endswith('.jpg')][:50]
print(f"Found {len(image_files)} images to process.")
```

```python
for img_name in image_files:
    input_image = os.path.join(INPUT_DIR, img_name)
    base_name = os.path.splitext(img_name)[0]
    cropped_image = os.path.join('test_images', f"{base_name}_cropped.png")
    bg_image = os.path.join('test_images', f"{base_name}_bg.png")

    print(f"\n{'='*40}\nProcessing {img_name}...\n{'='*40}")
    
    # 1. Preprocess
    print("Step 1: Preprocess (Crop Face)")
    cmd1 = f"{PYTHON_PATH} utils/preprocess.py --img_path {input_image} --save_path {cropped_image}"
    subprocess.run(cmd1, shell=True)

    # 2. Retrieve Background
    print("Step 2: Retrieve Background")
    cmd2 = f"{PYTHON_PATH} utils/retrieve_bg.py --img_path {input_image} --save_path {bg_image}"
    subprocess.run(cmd2, shell=True)

    # 3. Inference (Angry Face)
    print("Step 3: Inference (Angry)")
    cmd3 = f"{PYTHON_PATH} inference.py --img_path {cropped_image} --bg_path {bg_image} --au_test AU4+AU5 --AU_variation 5+5 --saved_path {OUTPUT_DIR}"
    subprocess.run(cmd3, shell=True)
    
    print(f"Finished processing {img_name}\n")

print("\nBatch processing complete!")
```
