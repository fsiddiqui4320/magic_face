# MagicFace Pipeline for WM_5

This notebook runs the complete MagicFace pipeline (preprocess, retrieve_bg, and inference) on the `WM_5` image.

```python
import os
import sys

# Add project root to path if needed
PROJECT_ROOT = os.getcwd()
if PROJECT_ROOT not in sys.path:
    sys.path.append(PROJECT_ROOT)

PYTHON_PATH = r'C:\Users\faris3\AppData\Local\miniconda3\envs\magicface\python.exe'
INPUT_IMAGE = os.path.join('test_images', 'processed_identities', 'WM_5.jpg')
CROPPED_IMAGE = os.path.join('test_images', 'WM_5_cropped.png')
BG_IMAGE = os.path.join('test_images', 'WM_5_bg.png')
OUTPUT_DIR = 'edited_images'

print(f"Project Root: {PROJECT_ROOT}")
print(f"Processing: {INPUT_IMAGE}")
```

## Step 1: Preprocess (Crop Face)
Crops the face from the original image and resizes to 512x512.

```python
!{PYTHON_PATH} utils/preprocess.py --img_path {INPUT_IMAGE} --save_path {CROPPED_IMAGE}
```

## Step 2: Retrieve Background
Generates the background mask and face landmarks.

```python
!{PYTHON_PATH} utils/retrieve_bg.py --img_path {INPUT_IMAGE} --save_path {BG_IMAGE}
```

## Step 3: Inference (Face Editing)
Applies facial expression changes (AU) to the image. Testing AU4+AU5 (Angry).

```python
!{PYTHON_PATH} inference.py --img_path {CROPPED_IMAGE} --bg_path {BG_IMAGE} --au_test AU4+AU5 --AU_variation 5+5 --saved_path {OUTPUT_DIR}
```

## Step 4: Display Results

```python
from PIL import Image
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 6))

# Cropped Face
plt.subplot(1, 3, 1)
if os.path.exists(CROPPED_IMAGE):
    plt.imshow(Image.open(CROPPED_IMAGE))
    plt.title("Cropped Face")
plt.axis('off')

# Background + Landmarks
plt.subplot(1, 3, 2)
if os.path.exists(BG_IMAGE):
    plt.imshow(Image.open(BG_IMAGE))
    plt.title("Background + Landmarks")
plt.axis('off')

# Edited Result
plt.subplot(1, 3, 3)
output_file = os.path.join(OUTPUT_DIR, os.path.basename(CROPPED_IMAGE))
if os.path.exists(output_file):
    plt.imshow(Image.open(output_file))
    plt.title("Edited Result (Angry)")
else:
    print(f"Waiting for {output_file}...")
plt.axis('off')

plt.show()
```
